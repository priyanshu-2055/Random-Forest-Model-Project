```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3,random_state=17)
```

```python
categorical=[var for var in xtrain.columns if xtrain[var].dtype=='O']
```

```python
categorical
```

```python
xtrain[categorical].isnull().sum()
```

```python
nan_catg=[var for var in xtrain[categorical] if xtrain[var].isnull().sum()>0]
nan_catg
```

```python
for df2 in [xtrain,xtest]:
  for var in nan_catg:
    df2[var].fillna(df2[var].mode()[0],inplace=True)
```

```python
xtrain[categorical].isnull().sum()
```

```python
xtest[categorical].isnull().sum()
```

```python
encoders=ce.OneHotEncoder(cols=categorical)
```

```python
xtrain=encoders.fit_transform(xtrain)
xtest=encoders.transform(xtest)
```

```python
cols=xtrain.columns
```

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
cols.shape
```

```python
xtrain=pd.DataFrame(xtrain,columns=[cols])
xtest=pd.DataFrame(xtest,columns=[cols])
```

```python
xtest.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=80,min_samples_split=6,oob_score=True)
rf.fit(xtrain,ytrain)
```

```python
ypred=rf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(classification_report(ytest,ypred))
accuracy_score(ytest,ypred)
```

```python
rf.oob_score_
```

```python
importance=rf.feature_importances_
```

```python
importance.shape
```

```python
xtrain.columns.shape,importance.shape
```

```python
feat_imp_df=pd.DataFrame({'features':xtrain.columns,'importance':importance})
feat_imp_df.head()
```

```python
feat_imp_df['features']=feat_imp_df['features'].apply(lambda feat:feat[0])
```

```python
feat_imp_df['features']
```

```python
plt.figure(figsize=(10,6))
plt.barh(feat_imp_df['features'],feat_imp_df['importance'])
plt.xlabel('importance')
plt.ylabel('features')
plt.show()
```

#Gridsearch

```python
n_estimators=[60,80,100]
max_depth=[None,2,4,6]
min_samples_split=[2,4,6]
para_grid={'n_estimators':n_estimators,'max_depth':max_depth,'min_samples_split':min_samples_split}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_search=RandomizedSearchCV(rf,para_grid,cv=5,scoring='accuracy')
```

```python
random_search.fit(xtrain,ytrain)
```

```python
random_search.best_params_
```

```python
rf2=RandomForestClassifier(n_estimators=100,min_samples_split=4,oob_score=True)
```

```python
rf2.fit(xtrain,ytrain)
```

```python
feature_score=rf2.feature_importances_
```

```python
feature_score.shape
```

```python
feature_importance=pd.DataFrame({'features':cols,'importance':feature_score})
feature_importance.sort_values(by='importance',inplace=True)
feature_importance.shape
```

```python
plt.figure(figsize=(15,6))
plt.bar(feature_importance['features'],feature_importance['importance'])
plt.xlabel('features')
plt.ylabel('importance')
plt.show()
```

```python
threshold=10**-4
sel_feat_imp=feature_importance[feature_importance['importance']>threshold]
```
